```python
import pandas as pd
import seaborn as sns 

df = pd.read_csv('data/q3.csv')

df['slot-datetime'] = pd.to_datetime(df['IOTERMIN'], infer_datetime_format=True)
df['slot-date'] = df['slot-datetime'].dt.date

counts = df.groupby("slot-date").size()
sns.distplot(counts, bins=100)
```

```python
import matplotlib.pyplot as plt
plt.show()
```

```python
df['slot-hour'] = df['slot-datetime'].dt.hour
```

```python
time_counts = df.groupby("slot-hour").size()
```

```python
sns.distplot(time_counts, bins=50, hist=True)
plt.show()
```

```python
import numpy as np
x = np.sort(df['slot-hour'].dropna().unique())
y = df.groupby('slot-hour').size()
```

```python
sns.lineplot(y.index, y)
```

```python
df['slot-month'] = df['slot-datetime'].dt.month
```

```python
x = np.sort(df['slot-month'].dropna().unique())
y = df.groupby('slot-month').size()
```

```python
sns.lineplot(y.index, y)
```
